```
np.random.seed(1254) ## For reproducibility

df['WEATHER'] = np.random.randint(1,4, size=len(df)) ## creates a random column of numbers 1,2,3

## create conditions list 
condition = [
             df['WEATHER'] == 1,
             df['WEATHER'] == 2,
             df['WEATHER'] == 3
             ]

#create output list for above conditions
output = [ 'Sun', 'Rain', 'Dry']


df['WEATHER'] = np.select(condition, output, default=np.nan)
```

```
np.random.seed(1254) 

df['WEATHER_2'] = np.random.randint(1,4, size=len(df))

df.loc[df['WEATHER_2'] == 1, :] = 'Sun'
df.loc[df['WEATHER_2'] == 2, :] = 'Rain'
df.loc[df['WEATHER_2'] == 3, :] = 'Dry'
```
